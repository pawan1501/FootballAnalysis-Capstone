```python
# 1. Check summary info to understand missing data and data types:
df.info()

# 2. Remove columns that have too many missing values (e.g., more than 50%)
threshold = len(df) * 0.5
df = df.dropna(axis=1, thresh=threshold)

# 3. Fill missing values in important numeric columns with 0 as default:
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(0)

# 4. For categorical columns, fill missing values with a placeholder string:
categorical_cols = df.select_dtypes(include='object').columns
df[categorical_cols] = df[categorical_cols].fillna('Unknown')

# 5. Convert date columns to datetime format (example for 'date_x'):
df['date_x'] = pd.to_datetime(df['date_x'], errors='coerce')

# 6. Remove duplicate rows if any:
df = df.drop_duplicates()

# 7. Reset index after cleaning:
df = df.reset_index(drop=True)

# 8. Optional: Rename columns to remove spaces or make consistent:
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Model 1: Simple Linear Regression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
df_simple = df[['goals', 'minutes_played']].dropna()
X_simple = df_simple[['minutes_played']]
y_simple = df_simple['goals']
X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(X_simple, y_simple, test_size=0.3, random_state=42)
lr_simple = LinearRegression()
lr_simple.fit(X_train_simple, y_train_simple)
y_pred_simple = lr_simple.predict(X_test_simple)
print("Simple Linear Regression")
print("Coefficients:", lr_simple.coef_)
print("Intercept:", lr_simple.intercept_)
print("Mean Squared Error:", mean_squared_error(y_test_simple, y_pred_simple))
print("R^2 Score:", r2_score(y_test_simple, y_pred_simple))
```

```python
# Model 2: Multiple Linear Regression
features = ['minutes_played', 'assists', 'yellow_cards', 'red_cards']
df_multi = df[['goals'] + features].dropna()
X_multi = df_multi[features]
y_multi = df_multi['goals']
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.3, random_state=42)
lr_multi = LinearRegression()
lr_multi.fit(X_train_multi, y_train_multi)
y_pred_multi = lr_multi.predict(X_test_multi)
print("\nMultiple Linear Regression")
print("Coefficients:", dict(zip(features, lr_multi.coef_)))
print("Intercept:", lr_multi.intercept_)
print("Mean Squared Error:", mean_squared_error(y_test_multi, y_pred_multi))
print("R^2 Score:", r2_score(y_test_multi, y_pred_multi))
```
